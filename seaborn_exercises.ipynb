4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python

chipotle_sql_query = '''
                     SELECT *
                     FROM orders;
                     '''
```

```python
orders = pd.read_sql(chipotle_sql_query, get_db_url('chipotle'))
orders.head()
```

```python
orders['item_price'] = orders.item_price.str.replace('$', '').astype(float)
orders.info()
```

Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
top_four = orders.groupby('item_name').quantity.sum().sort_values(ascending=False).head(4)]
top_four
```

```python
top_four = top_four.reset_index()
```

```python
type(top_four)
```

```python
top_four
```

```python
top_four.columns
```

```python
top_rev = orders.groupby('item_name').item_price.sum()
```

```python
top_rev = top_rev.reset_index()
```

```python
top_four_popolar_rev=top_rev.merge(top_four , 
            left_on='item_name', 
            right_on='item_name', 
            how='right')
top_four_popolar_rev
```

```python
top_four_popolar_rev.columns
```

```python
sns.barplot(data = top_four_popolar_rev, x = 'item_price',y = 'item_name')
```

5. Load the sleepstudy data and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.

```python
sleepstudy = data('sleepstudy')
```

```python
sleepstudy
```

```python
sleepstudy['Subject'] = 'Subject_' + sleepstudy.Subject.astype(str) 
```

```python
plt.figure(figsize=(16, 9))

sns.lineplot(data = sleepstudy, x = 'Days', y = 'Reaction', hue = 'Subject', legend=False)
sns.lineplot(data = sleepstudy, x = 'Days', y = 'Reaction', color = 'black', linestyle='--', estimator = 'mean')
plt.annotate('Average',(0,265),fontsize=17)
```
